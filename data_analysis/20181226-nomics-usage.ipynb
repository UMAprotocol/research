```python
import requests
import datetime as dt
import pandas as pd
import numpy as np
import os
import json

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
BASEURL = "https://api.nomics.com/v1"

def get_quote_rates(APIKEY):
    if not os.path.exists("../nomics_rates"):
        os.mkdir("../nomics_rates")
        
    get_url = BASEURL + "/exchange-rates"
    
    _params = {
        "key": APIKEY
    }

    response = requests.get(get_url, _params)
    response_json = response.json()
    df = pd.DataFrame(response_json)
    df.loc[:, 'timestamp'] = pd.to_datetime(df['timestamp'])
    
    return df

def exists_in_currencies(APIKEY, target):
    if not os.path.exists("../nomics_rates"):
        os.mkdir("../nomics_rates")
        
    get_url = BASEURL + "/currencies"
    
    _params = {
        "key": APIKEY
    }

    response = requests.get(get_url, _params)
    response_json = response.json()
    df = pd.DataFrame(response_json)
    
    exists = target in df
    
    return exists

def load_historical_price(APIKEY, start_time, end_time):
    
    get_url = BASEURL + "/currencies/interval"
        
    _params = {
        "key": APIKEY,
        "start": start_time,
        "end": end_time,
    }

    response = requests.get(get_url, params=_params)
    response_json = response.json()
    df = pd.DataFrame(response_json)
    return df
```

```python

apikey = json.loads(open("../nomics_apikey.json").read())["APIKEY"]
# list_quote_rates = get_quote_rates(apikey)
# does_DAI_exist = exists_in_currencies(apikey, "DAI")

fmt_dt = lambda d: "{:%Y-%m-%dT%H:%M:%SZ}".format(d)
start_dt = dt.datetime(2018, 1, 1)
end_dt = dt.datetime(2018, 12, 31)
hist_prices = load_historical_price(apikey, fmt_dt(start_dt), fmt_dt(end_dt))
```

```python
asset_to_view = "DAI"
data = hist_prices.pivot(index = "close_timestamp", columns = 'currency')['close']
data_asset = data.loc[:,asset_to_view]

print(data_asset.dropna())
```
